```python
import pandas as pd
```

```python
df1=pd.read_csv('AWCustomers.csv')
df2=pd.read_csv('AWSales.csv')
```

```python
df1.info()
```

```python
df2.info()
```

```python
df_merged=pd.merge(df1,df2,left_on='CustomerID',right_on='CustomerID',how='inner')
df_merged.head()
```

```python
df_merged.info()
```

```python
df_merged.isnull().sum()
```

```python
df_merged['CustomerID'].unique()
```

```python
df_merged=df_merged.drop_duplicates(subset='CustomerID', keep='first')
```

```python
df_merged.info()
```

```python
columns_to_drop = [
    'CustomerID', 'Title', 'FirstName', 'MiddleName', 'LastName', 'Suffix',
    'AddressLine1', 'AddressLine2', 'PhoneNumber', 'BirthDate','PostalCode','TotalChildren','LastUpdated'
]
df_dropped = df_merged.drop(columns=columns_to_drop)
```

```python
df_dropped.head()
```

```python
df_dropped.isnull().sum()
```

```python
df_dropped.info()
```

```python
for col in df_dropped.select_dtypes(include='object').columns:
    print(f"Unique values in '{col}':")
    print(df_dropped[col].unique())
    print("-" *40)
```

```python
df_dropped['Gender']=df_dropped['Gender'].map({
    'M':1,
    'F':0
})
df_dropped['MaritalStatus']=df_dropped['MaritalStatus'].map({
    'M':1,
    'S':0
})
```

```python
for col in df_dropped.select_dtypes(include='object').columns:
    print(f"Unique values in '{col}':")
    print(df_dropped[col].unique())
    print("-" *40)
```

```python
education_order = {
    'Partial High School': 0,
    'High School': 1,
    'Partial College': 2,
    'Bachelors': 3,
    'Graduate Degree': 4
}
df_dropped['Education'] = df_dropped['Education'].map(education_order)
```

```python
import pandas as pd

# Assuming your DataFrame is called df
Q1 = df_dropped['YearlyIncome'].quantile(0.25)
Q3 = df_dropped['YearlyIncome'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identify outliers
outliers = df_dropped[(df_dropped['YearlyIncome'] < lower_bound) | (df_dropped['YearlyIncome'] > upper_bound)]

print(f"Number of outliers: {len(outliers)}")
```

```python
df_encoded = pd.get_dummies(df_dropped, columns=['Occupation', 'CountryRegionName'])
df_encoded.head()
```

```python
df_encoded.describe()
```

```python
X = df_encoded.drop(columns=['StateProvinceName', 'City', 'BikeBuyer'])
y = df_encoded['BikeBuyer']
```

```python
X.head()
y.head()
```

```python
from sklearn.preprocessing import RobustScaler

scaler = RobustScaler()
X[['YearlyIncome', 'AvgMonthSpend']] = scaler.fit_transform(X[['YearlyIncome', 'AvgMonthSpend']])
X[['YearlyIncome', 'AvgMonthSpend']].head()
```

```python
X.describe()
```

```python
y.value_counts()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test=train_test_split(X,y,test_size=0.2,stratify=y)
```

```python
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
for col in df_encoded.select_dtypes(include='object').columns:
    print(f"Unique values in '{col}':")
    print(df_dropped[col].unique())
    print("-" *40)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf_model=RandomForestClassifier()
rf_model.fit(X_train,y_train)
rf_model.score(X_test,y_test)
```

```python
from sklearn.metrics import classification_report
y_predrf=rf_model.predict(X_test)
print(classification_report(y_test,y_predrf))
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn_model=KNeighborsClassifier()
knn_model.fit(X_train,y_train)
print(knn_model.score(X_test,y_test))
from sklearn.metrics import classification_report
y_predknn=knn_model.predict(X_test)
print(classification_report(y_test,y_predknn))
```

```python
from sklearn.ensemble import GradientBoostingClassifier
gb_model = GradientBoostingClassifier()
gb_model.fit(X_train, y_train)

print("Gradient Boosting Test Accuracy:", gb_model.score(X_test, y_test))

from sklearn.metrics import classification_report
y_pred_gb = gb_model.predict(X_test)
print(classification_report(y_test, y_pred_gb))
```

```python
from sklearn.ensemble import AdaBoostClassifier

ada_model = AdaBoostClassifier()
ada_model.fit(X_train, y_train)

print("AdaBoost Test Accuracy:", ada_model.score(X_test, y_test))

from sklearn.metrics import classification_report
y_pred_ada = ada_model.predict(X_test)
print(classification_report(y_test, y_pred_ada))
```

```python
from xgboost import XGBClassifier
from sklearn.metrics import classification_report

# Initialize and fit the model
xgb_model = XGBClassifier(use_label_encoder=False, eval_metric='logloss')  # suppress warnings
xgb_model.fit(X_train, y_train)

# Test accuracy
print("XGBoost Test Accuracy:", xgb_model.score(X_test, y_test))

# Classification report
y_pred_xgb = xgb_model.predict(X_test)
print(classification_report(y_test, y_pred_xgb))
```
